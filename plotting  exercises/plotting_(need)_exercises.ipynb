**Exercise 1**

```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',   # we can change the marker and marker colour for error here
             ecolor='black',capsize=5,     # ecolour is to change the error bar colour and capsize change the size
             label='Y values with errors') 
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

**Exercise 2**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 50, 200)  
y = np.exp(-x/10)*np.sin(x)

plt.plot(x, y)

plt.xlabel('x', fontsize=15, color='darkorange')  # Set fontsize and color for x-label
plt.ylabel('y', fontsize=15, color='darkorange', rotation=0)  # Set fontsize and color for y-label
                                                            # Note that i also set the rotation to 0 so that the label look upright
plt.title(r'$y = e^{-x/10} \sin(x)$', fontsize= 20)  # Note here!!! I tried for very long. Use r string to express this and use two money sign to make 
                                                    # it a math equation
plt.grid(alpha=0.25)  # Set grid opacity

plt.ylim([-1,1])
```
