# What to expect in this chapter

So far, the ‘dialect’ we have used to talk to Matplotlib is called the **Matlab-like pyplot(plt) interface**. From here onward, I will show you how to use the other, more powerful ‘dialect’ called the **Object Oriented (OO) interface**. This way of talking to Matplotlib gives us a more nuanced control over what is going on by allowing us to manipulate the various axes easily.

# 1 Some nomenclature

- **Figure** is the full canvas you use to draw stuff on
- **Axis** is the individual math axis(a set of y-x acis for example)

# 2 Comparing the two ‘dialects’

```python
import numpy as np
from matplotlib import pyplot as plt

# generating data
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)

#This is the way we learned using pyplot
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()

# This is the new way using oo interface
fig, ax = plt.subplots(nrows=1, ncols=1) #plt.subplots() is a function that creates a new figure and a set of subplots.
                                        #It returns a tuple containing a reference to the figure (fig) and an array of axes (ax). 
                                        #nrows=1 and ncols=1 specify that we want a single subplot
ax.plot(x, cos_x, label='cos x') #ax.plot() is a method used to plot data on the axes
ax.plot(x, sin_x, label='sin x')
ax.legend()
```

# 3 What is OO ax, really?

```python
fig, ax = plt.subplots(nrows=2, ncols=1)  # here we ask for two rows and one column, this creates 2 axis

ax[0].plot(x, cos_x, label='cos x') # axis 0 to access the first axis to plot on
ax[1].plot(x, sin_x, label='sin x') # axis 1 to access the second axis to plot on

ax[0].legend()
ax[1].legend()
```

**Let's look at the properties of the axis**

```python
type(ax) # what is ax? It is an numpy array!
```

```python
ax.shape # what size is ax? There are two things in this figure above!
```

```python
type(ax[0]) # what is contained in ax? it is a matplotlib axis.
```

# 4 A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=1,  #Get two axes ready to plot in two rows in one column
                       figsize=(5, 5),  #Change the size of the figure by specifying a figure size (figsize)
                       sharex=True)  # Ask that the plots share the x-axis using sharex

ax[0].plot(x, cos_x, label=r'$\cos(x)$')  #Since ax is a NumPy array with two axes, we can index them using 0 and 1. Then we just plot to that axis.
ax[0].fill_between(x, 0, cos_x, alpha=.25) # we see that there are 4 arguments here
                                            # x means that we want to fill all the values of x with colours. 0 is a bound and cos_x is another bound
                                           # In this case, we are saying that fill the area between y=0 and the curve with colour

ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:  # we loop here for both axis
    a.legend()  # show the legend
    a.grid(alpha=.25)  # draw the grid with 0.25 alpha
    a.set_ylabel('$y$', rotation=0)  # give y a label of 'y' that is upright

ax[1].set_xlabel('$x$')  # for the second axis we give it a x axis label

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')  # give title
fig.tight_layout() #make any necessary adjustments to the layout to make our plot look nice by calling tight_layout()
```
